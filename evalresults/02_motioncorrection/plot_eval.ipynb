```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

# load data
# ssim metrics
ssim_ants = pd.read_csv(r'evalresults/ssim_mc_ants.csv', usecols=["ants (rigid)", "ants (affine)", "ants (deform)"])
ssim_niftyreg = pd.read_csv(r'evalresults/ssim_mc_niftyreg.csv', usecols=["niftyreg (5x)", "niftyreg (default)", "niftyreg (10x)"])
ssim_elastix = pd.read_csv(r'evalresults/ssim_mc_elastix.csv', usecols=["elastix"])

# runtimes
rt_ants = pd.read_csv(r'evalresults/runtime_mc_ants.csv', usecols=["ants (rigid)", "ants (affine)", "ants (deform)"])
rt_niftyreg = pd.read_csv(r'evalresults/runtime_mc_niftyreg.csv', usecols=["niftyreg (5x)", "niftyreg (default)", "niftyreg (10x)"])
rt_elastix = pd.read_csv(r'evalresults/runtime_mc_elastix.csv', usecols=["elastix"])

ssim = pd.concat([ssim_ants, ssim_niftyreg, ssim_elastix], axis=1)
rt = pd.concat([rt_ants, rt_niftyreg, rt_elastix], axis=1)
```

```python
# [ants rigid, ants affine, ants deform, niftyreg 5x, niftyreg default, niftyreg 10x, elastix]
# [light blue, blue, dark blue, green, teal, green(er), purple]
alg_cols = ['#9796d1', '#5250b7', '#1a1887', '#23ea47', '#23eaab', '#62ea23', '#ad37ad']
```

```python
plt.rcParams["figure.figsize"] = (12, 9) # (w, h)

sns.boxplot(data=rt, palette=alg_cols, linewidth=2, saturation=0.7)
plt.title('Processing times for motion-correction', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Algorithm/Program Package', fontsize=12)
plt.ylabel('Mean Run-Time (s)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
# Adding a background color for the plot
plt.gca().set_facecolor('#f0f0f5')
```

```python
plt.rcParams["figure.figsize"] = (12, 9) # (w, h)

# Violin plot with customizations
sns.violinplot(data=ssim, palette=alg_cols, linewidth=1.5, saturation=0.9)
plt.title('SSIM metric (SSIMM) for Motion-Correction', fontsize=18, fontweight='bold', color='darkred')
plt.xlabel('Program Package/Algorithm', fontsize=12)
plt.ylabel('SSIMM', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# Adding a background color for the plot
plt.gca().set_facecolor('#f9f9f9')

# Adjust layout to avoid overlap
plt.tight_layout()
```
