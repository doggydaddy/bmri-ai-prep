```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

# load data
# ssim metrics
ssim_coreg = pd.read_csv(r'evalresults/ssim_coreg.csv')
print(ssim_coreg)
```

```python
# [ants rigid, ants affine, ants deform, niftyreg 5x, niftyreg default, niftyreg 10x, elastix]
# [light blue, blue, dark blue, green, teal, green(er), purple]
alg_cols = ['#9796d1', '#5250b7', '#1a1887', '#23ea47', '#23eaab', '#62ea23', '#ad37ad']
```

```python
plt.rcParams["figure.figsize"] = (12, 9) # (w, h)

sns.boxplot(data=ssim_coreg, linewidth=2, saturation=0.7)
plt.title('SSIM metric (SSIMM) for co-registration', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Regularization Parameters', fontsize=12)
plt.ylabel('SSIMM', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
# Adding a background color for the plot
plt.gca().set_facecolor('#f0f0f5')
```

```python
plt.rcParams["figure.figsize"] = (12, 9) # (w, h)

# Violin plot with customizations
sns.violinplot(data=ssim_coreg, linewidth=1.5, saturation=0.9)
plt.title('SSIM metric (SSIMM) for Co-Registration', fontsize=18, fontweight='bold', color='darkred')
plt.xlabel('Regularization Parameters', fontsize=12)
plt.ylabel('SSIMM', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# Adding a background color for the plot
plt.gca().set_facecolor('#f9f9f9')

# Adjust layout to avoid overlap
plt.tight_layout()
```
