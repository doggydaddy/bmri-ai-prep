```python
# cell to evaluate ANTs rigid, affine, and deformation algorithms

import os, re
import time

# Replace '/path/to/your/directory' with the actual root directory you want to search
root_directory = '../evaldata'
match_string = 't1.nii'  # Replace with the string you're looking for

files = find_files_matching_string(root_directory, match_string)

n = len(files)
c = 0
ssim_results = np.zeros((n, 6))
for file in files:

    print("processing: ", file)

    # we split input filenames so we can easily name the output filenames 
    dirname, basename = os.path.split(file)

    split_filename = file.split("/")
    subj = split_filename[2]
    sess = split_filename[3]

    ref_name = re.sub("t1", "ref", basename)
    ref = os.path.join(dirname, ref_name)

    output_dir = r'evalresults'

    output_name_1 = os.path.join(output_dir, re.sub("t1", "t1_RegStd", basename))
    output_name_2 = re.sub("t1", "t1_Reg10x", basename)
    output_name_3 = re.sub("t1", "t1_Reg20x", basename)
    output_name_4 = re.sub("t1", "t1_Reg30x", basename)
    output_name_5 = re.sub("t1", "t1_Reg40x", basename)
    output_name_6 = re.sub("t1", "t1_Reg50x", basename)
    output_cpp_1 = os.path.join(output_dir, re.sub("t1.nii", "t1_RegStd.cpp", basename))
    output_cpp_2 = re.sub("t1.nii", "t1_Reg10x.cpp", basename)
    output_cpp_3 = re.sub("t1.nii", "t1_Reg20x.cpp", basename)
    output_cpp_4 = re.sub("t1.nii", "t1_Reg30x.cpp", basename)
    output_cpp_5 = re.sub("t1.nii", "t1_Reg40x.cpp", basename)
    output_cpp_6 = re.sub("t1.nii", "t1_Reg50x.cpp", basename)

    coreg_nrdef(ref, file, output_cpp_1, output_name_1, be='0.001', le='0.01')
    coreg_nrdef(ref, file, output_cpp_2, output_name_2, be='0.01', le='0.1')
    coreg_nrdef(ref, file, output_cpp_3, output_name_3, be='0.02', le='0.2')
    coreg_nrdef(ref, file, output_cpp_4, output_name_4, be='0.03', le='0.3')
    coreg_nrdef(ref, file, output_cpp_5, output_name_5, be='0.04', le='0.4')
    coreg_nrdef(ref, file, output_cpp_6, output_name_6, be='0.05', le='0.5')

    ssim_results[c, 0] = eval_coreg_ssim(ref, file, output_name_1)
    ssim_results[c, 1] = eval_coreg_ssim(ref, file, output_name_2)
    ssim_results[c, 2] = eval_coreg_ssim(ref, file, output_name_3)
    ssim_results[c, 3] = eval_coreg_ssim(ref, file, output_name_4)
    ssim_results[c, 4] = eval_coreg_ssim(ref, file, output_name_5)
    ssim_results[c, 5] = eval_coreg_ssim(ref, file, output_name_6)
    c += 1

import pandas as pd
cn = ["Default", "10x", "20x", "30x", "40x", "50x"]
ssim_coreg = pd.DataFrame(ssim_results, columns=cn)
ssim_coreg.to_csv(r'evalresults/ssim_coreg.csv', index=False)
```
